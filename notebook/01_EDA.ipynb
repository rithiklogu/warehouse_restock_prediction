```python
df = df.sort_values(
    by=["product_id", "hub_id", "year", "month"]
).reset_index(drop=True)
```

```python
df
```

```python
group = df.groupby(
    ["product_id", "hub_id"]
)
```

```python
group
```

```python
df["sales_last_1"] = group["quantity_sold"].shift(1)

df["sales_last_2"] = group["quantity_sold"].shift(2)

df["sales_last_3"] = group["quantity_sold"].shift(3)
```

```python
df["avg_sales_3"] = df[
    [
        "sales_last_1",
        "sales_last_2",
        "sales_last_3"
    ]
].mean(axis=1)
```

```python
df["target_next_month_demand"] = (
    group["quantity_sold"]
    .shift(-1)
)
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df
```

```python
training_df = df[
    [
        "product_id",
        "hub_id",
        "year",
        "month",
        "current_stock",
        "sales_last_1",
        "sales_last_2",
        "sales_last_3",
        "avg_sales_3",
        "promotion",
        "festival_flag",
        "discount_percentage",
        "returns",
        "target_next_month_demand",
    ]
]
```

```python
training_df
```

```python
print(training_df.head())

print(training_df.shape)

print(training_df.isnull().sum())
```

```python
training_df["date"] = pd.to_datetime(
    training_df["year"].astype(str)
    + "-"
    + training_df["month"].astype(str)
    + "-01"
)
```

```python
train_df = training_df[
    training_df["date"] < "2025-01-01"
]
```

```python
val_df = training_df[
    (training_df["date"] >= "2025-01-01") &
    (training_df["date"] < "2025-07-01")
]
```

```python
test_df = training_df[
    training_df["date"] >= "2025-07-01"
]
```

```python
X_train = train_df.drop(
    columns=["target_next_month_demand", "date"]
)

y_train = train_df["target_next_month_demand"]
```

```python
X_val = val_df.drop(
    columns=["target_next_month_demand", "date"]
)

y_val = val_df["target_next_month_demand"]
```

```python
X_test = test_df.drop(
    columns=["target_next_month_demand", "date"]
)

y_test = test_df["target_next_month_demand"]
```

```python
print("Train :", X_train.shape)
print("Validation :", X_val.shape)
print("Test :", X_test.shape)
```

```python
from catboost import CatBoostRegressor

cat_features = ["product_id", "hub_id"]

model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.05,
    depth=8,
    loss_function="RMSE",
    random_seed=42,
    verbose=100
)

model.fit(
    X_train,
    y_train,
    cat_features=cat_features,
    eval_set=(X_val, y_val),
    use_best_model=True
)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("MAE :", mae)
print("RMSE:", rmse)
print("R²  :", r2)
print("MAPE:", mape)
```

```python
feature_importance = model.get_feature_importance()

importance_df = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": feature_importance
}).sort_values("Importance", ascending=False)

print(importance_df)
```
